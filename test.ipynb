```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# present the ideal case when d=N
N_k_d_s = [2, 4, 6, 8, 10]
length = len(N_k_d_s)
data_csv = pd.read_csv("resluts/scores_10.csv").values[:, 1:]
data = data_csv.reshape(length, length, length)
#print(data)
ideal_scores = [data[i][i][i] for i in range(length)]
plt.scatter(N_k_d_s, ideal_scores)
plt.ylim((0, 1.05))
plt.xlabel("N_k")
plt.ylabel("MRR")
plt.title("the ideal(trivail) case!")
```

```python
# when d achieve the goal scores fixed k/N = 1
k_N = 1
goal = 0.9
collect = []
[int(k_N*i) +1 for i in range(length)]
for i in range(length):
    base_score = 0
    for d in range(int(k_N*i)+1):
        if data[i][-1][int(d)] > base_score:
            base_score = data[i][-1][int(d)]
        if base_score > goal:
            break
#        print(data[i][-1][int(d)])
    collect.append([base_score, 2*(d+1)/N_k_d_s[i]])
collect = np.array(collect)
plt.scatter(N_k_d_s, collect[:,0], alpha=0.5)
plt.scatter(N_k_d_s, collect[:,1], alpha=0.5)
plt.title("when d achieve the goal scores fixed k/N = 1")
plt.legend(["MRR", "d / N"])
plt.xlabel("N_k")
```

```python
# how d perform  whens k is fixed and less than N.(k=4)
k_N = 1
k = 1
collect = []
[int(k_N*i) +1 for i in range(length)]
for i in range(length):
    plt.scatter(N_k_d_s[i:], (data[i:, k, i]))
plt.legend(["d=2", "d=4", "d=6", "d=8", "d=10"])
plt.ylabel("MRR")
plt.xlabel("N")
plt.title("how d perform  whens k is fixed and less than N.(k=4)")
```

```python
# how d perform  whens k and N are fixed. (k=2, N=10)
#The results show that the score would be better with the improvment of d.
plt.scatter(N_k_d_s, data[-1][0])
plt.ylabel("MRR")
plt.xlabel("d")
```

```python
# how k perform  whens N and d is changing(N=10, d=4).
# the results shows that the score may imporve when more sets to be trained. 
plt.scatter(N_k_d_s, data[-1, :, 1])
plt.ylabel("MRR")
plt.xlabel("k")
```
